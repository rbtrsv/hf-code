```python
data_analysts = {'Name':['Robert', 'Andrei', 'Claudiu'], 
                 'Experience':[2, 3, 5],
                 'Music': ['techno', '3sudest', 'rock'],
                 'Progress': [1.134, 1.143, 1.145]}
df_analysts = pd.DataFrame(data_analysts)
df_analysts
```

```python
def xlookup(lookup_value, lookup_array, return_array, if_not_found:str = ''):
    match_value = return_array.loc[lookup_array == lookup_value]
    if match_value.empty:
        return f'"{lookup_value}" not found!' if if_not_found == '' else if_not_found

    else:
        return match_value.tolist()[0]
```

```python
who = xlookup('techno', df_analysts['Music'], df_analysts['Name'])
who
```

```python
df_analysts['Productivity'] = df_analysts['Experience'] * df_analysts['Progress']
df_analysts
```

```python
random_ratio = 2.45
df_analysts['Acutal Productivity'] = df_analysts['Productivity'].apply(lambda x: x * random_ratio)
df_analysts
```
